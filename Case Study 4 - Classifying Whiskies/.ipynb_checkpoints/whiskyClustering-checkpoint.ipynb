## I. Loading and Inspecting Data

```python
import pandas as pd
import numpy as np
```

```python
whisky = pd.read_csv('whiskies.txt')
```

```python
whisky['Region'] = pd.read_csv('regions.txt')
```

```python
whisky.head()
```

```python
flavors = whisky.iloc[:, 2:14]
```

```python
flavors.head()
```

```python
flavors.columns
```

## II. Exploring Correlations

```python
corr_flavors = pd.DataFrame.corr(flavors)
```

```python
print(corr_flavors)
```

#### Visualising correlation between different flavor attributes

```python
# visualise the correlation matrix
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.pcolor(corr_flavors, cmap = 'Spectral')
plt.colorbar()
plt.show()
```

#### Correlation among whiskies across flavors

```python
corr_whisky = pd.DataFrame.corr(flavors.transpose())
```

```python
plt.figure(figsize=(10,10))
plt.pcolor(corr_whisky, cmap='Spectral')
plt.colorbar()
plt.show()
```

## III. Clustering Whiskies by Flavor Profile

```python
from sklearn.cluster import SpectralCoclustering
```

```python
model = SpectralCoclustering(n_clusters=6, random_state=0)
```

```python
model.fit(corr_whisky)
```

The output of the following command is an array with the following dimensions:
(number of row clusters) * (number of rows in the data matrix)

```python
# model.rows_
```

```python
# Number of observations belonging to each cluster.
np.sum(model.rows_, axis=1)
```

```python
np.sum(model.rows_, axis=0)
```

```python
model.row_labels_
```

## IV. Comparing Correlation Matrices

```python
whisky['Group'] = pd.Series(model.row_labels_, index=whisky.index)
```

```python
whisky = whisky.iloc[np.argsort(model.row_labels_)]
```

```python
whisky = whisky.reset_index(drop=True)
```

```python
whisky.head()
```

```python
correlations = pd.DataFrame.corr(whisky.iloc[:, 2:14].transpose())
```

```python
correlations = np.array(correlations)
```

```python
plt.figure(figsize = (14,7))
plt.subplot(121)
plt.pcolor(corr_whisky, cmap='Spectral')
plt.title("Original")
plt.axis("tight")
plt.subplot(122)
plt.pcolor(correlations, cmap='Spectral')
plt.title("Rearranged")
plt.axis("tight")
plt.show()
```
