```python
import numpy as np
```

```python
x = np.linspace(0, 10, 20)
```

```python
y1 = x**2.0
```

```python
y2 = x**3
```

```python
plt.loglog(x, y1, "ro-", linewidth=2, markersize=12, label="First")
plt.loglog(x, y2, "gd-", linewidth=2, markersize=12, label="Second")
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.legend(loc="upper left")
plt.axis([-0.5, 10.5, -5, 105])
```
