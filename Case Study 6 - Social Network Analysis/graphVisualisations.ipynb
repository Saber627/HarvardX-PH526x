```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.karate_club_graph()
```

```python
nx.draw(G, with_labels=True, node_color="lightblue", edge_color="grey")
```

```python
G.degree()
```

```python
G.degree()[33]
```

```python
G.degree(33)
```

Python distinguishes between the two calls based on their call signature. The former has no arguments, while the latter has one argument, i.e. the ID of the node whose degree we are trying to querying. This is how Python knows whether to return a dictionary or whether to use the function to obtain the degree of the given node.

```python
G.number_of_nodes()
```
