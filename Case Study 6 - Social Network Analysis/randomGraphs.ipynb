```python
import networkx as nx
from scipy.stats import bernoulli
```

I'm going to implement an Erdos-Renyi Graph model as a python function!

```python
# Step-1: Implementing coin-flip. This function will determine (randomly) the existence of an edge between a pair of nodes.
bernoulli.rvs(p=0.6)
```

1. Create empty graph
2. add all N nodes in the graph
3. loop over all pairs of nodes
   3.1 add an edge with prob p

```python
N = 20
p = 0.2
```

```python
G = nx.Graph()
G.add_nodes_from(range(N))
for node1 in G.nodes():
    for node2 in G.nodes():
        if (node1 < node2) and (bernoulli.rvs(p=p)):
            G.add_edge(node1, node2)
```

```python
F = nx.Graph()
F.add_nodes_from(range(N))
for node1 in F.nodes():
    for node2 in F.nodes():
        if (node1 < node2) and (bernoulli.rvs(p=p)):
            F.add_edge(node1, node2)
```

```python
nx.draw(F, with_labels=True)
```

```python
def er_graph(N, p):
    """Generate an ER graph."""
    G = nx.Graph()
    G.add_nodes_from(range(N))
    for node1 in G.nodes():
        for node2 in G.nodes():
            if (node1<node2) and (bernoulli.rvs(p=p)):
                G.add_edge(node1, node2)
    return G
```

```python
nx.draw(er_graph(50, 0.08), node_size=40, node_color='g')
```
